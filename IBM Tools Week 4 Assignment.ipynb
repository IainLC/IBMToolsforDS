### In the proceeding cell(5) I will demonstrate how one uses python to find the sum of an array

````python
def _sum(arr): 
    sum=0
    
    for i in arr:
        sum = sum + i
          
    return(sum) 
  
arr=[] 
arr = [563654, 1000, 78946, 52452456] 
   
n = len(arr) 
ans = _sum(arr) 
  
print ('Sum of the array is ', ans)
````

--- 
The Python code above includes
* A variable to initialize. ``` def_sum(arr)```
* A for loop to iterate through the array. ``` for i in arr: sum = sum + i return(sum)```
* A Driver function ```arr=[]```
* An array of integers ```arr = [1,2,3,4,..]```
* An Inbuilt function for caluculating the Sum
  * ```Sum()``` Adds all the elements in the lis, set and tuples and returns the value
